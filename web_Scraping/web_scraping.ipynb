Scraping till End Page

```python
def scrape_quotes(url):
    """Scrapes quotes, authors, and tags from a given URL and saves them to a CSV file."""

    try:
        page_to_scrape = requests.get(url)
        page_to_scrape.raise_for_status()  # Raise an exception for bad status codes

        soup = BeautifulSoup(page_to_scrape.text, 'html.parser')
        quotes = soup.findAll('span', attrs={'class': 'text'})
        authors = soup.findAll('small', attrs={'class': 'author'})
        tags = soup.findAll('div', attrs={'class': 'tags'})

        with open('quotes.csv', 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['Quote', 'Author', 'Tags'])  # Write header row

            for i in range(0, len(quotes)):
                quote_text = quotes[i].text.strip()
                author_text = authors[i].text.strip()
                quote_tags = [tag.text.strip() for tag in tags[i].findAll('a', attrs={'class': 'tag'})]
                writer.writerow([quote_text, author_text, ', '.join(quote_tags)])

        print("Quotes scraped and saved to quotes.csv")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Start with the first page
url = 'https://quotes.toscrape.com/'
scrape_quotes(url)


# Now loop through next pages if they exist
while True:
    next_page = soup.find('li', class_='next')
    if next_page:
        next_page_link = next_page.find('a')['href']
        url = 'https://quotes.toscrape.com/' + next_page_link # Correct url construction
        scrape_quotes(url)
    else:
        break
```
