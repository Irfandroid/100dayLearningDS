# Step 1: Importing Data
We will use the pandas library to manage data in a DataFrame

```python
import pandas as pd

# Contoh dataset sederhana
data = {
    'Name': ['Budi', 'Siti', 'Andi', 'Rina'],
    'Age': [22, 21, 23, 22],
    'Gender': ['L', 'P', 'L', 'P'],
    'Test_Score': [85.5, 90.0, 78.5, 88.0]
}

df = pd.DataFrame(data)
print(df)
```

# Step 2: Checking the Data Type
 Pandas allows us to check the data type of each column.

```python
print(df.dtypes)
```

The output will show the data type of each column, whether it is int, float, or object (for strings/categories).

# Step 3: Analyzing the Data
From this dataset, we can identify:

Age and Test_Score columns as numerical data.
Name and Gender columns as categorical data.

# Step 4: Data Visualization
We can visualize the data to make it easier to understand:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualisasi distribusi usia
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```
